```python
import pyodbc
```

```python
# Crez la connexion
connection = pyodbc.connect('DRIVER={SQL SERVER} ;SERVER=LAPTOP-1US3GU3J\SQLEXPRESS;DATABASE=GithubDB')
```

```python
cursor = connection.cursor()
```

```python
# Create the Contributor table
cursor.execute('''
    CREATE TABLE [Contributor] (
      [contributor_Id] INT,
      [contributor_name] VARCHAR(255),
      PRIMARY KEY ([contributor_Id])
    );
''')

# Create the Language table
cursor.execute('''
    CREATE TABLE [Language] (
      [language_Id] INT,
      [language_name] VARCHAR(255),
      PRIMARY KEY ([language_Id])
    );
''')

# Create the Owner table
cursor.execute('''
    CREATE TABLE [Owner] (
      [owner_Id] INT,
      [owner_name] VARCHAR(255),
      PRIMARY KEY ([owner_Id])
    );
''')

# Create the Repository table
cursor.execute('''
    CREATE TABLE [Repository] (
      [repository_Id] INT,
      [owner_Id] INT,
      [repository_name] VARCHAR(255),
      [repository_url] VARCHAR(255),
      [created_at] DATETIME,
      [description] VARCHAR(1000),
      [topics] VARCHAR(400),
      [language_Id] INT,
      [stars] INT,
      [forks] INT,
      [open_issues] INT,
      PRIMARY KEY ([repository_Id]),
      CONSTRAINT [FK_Repository.language_Id]
        FOREIGN KEY ([language_Id])
          REFERENCES [Language]([language_Id]),
      CONSTRAINT [FK_Repository.owner_Id]
        FOREIGN KEY ([owner_Id])
          REFERENCES [Owner]([owner_Id])
    );
''')

# Create the Repository_Contributor table
cursor.execute('''
    CREATE TABLE [Repository_Contributor] (
      [repository_Id] INT,
      [contributor_Id] INT,
      [contributions_count] INT,
      CONSTRAINT [FK_Repository_Contributor.contributor_Id]
        FOREIGN KEY ([contributor_Id])
          REFERENCES [Contributor]([contributor_Id]),
      CONSTRAINT [FK_Repository_Contributor.repository_Id]
        FOREIGN KEY ([repository_Id])
          REFERENCES [Repository]([repository_Id])
    );
''')
```

```python
connection.commit()
```

```python
# Insert data into the Contributor table
for row in contributor.itertuples():
    # Check if the contributor_Id already exists in the table
    cursor.execute('SELECT COUNT(*) FROM Contributor WHERE contributor_Id = ?', (row.contributor_Id,))
    count = cursor.fetchone()[0]
    
    if count == 0:
        # Insert the record if it doesn't exist
        cursor.execute('INSERT INTO Contributor (contributor_Id, contributor_name) VALUES (?, ?)', (row.contributor_Id, row.contributor))

connection.commit()
```

```python
# Insert data into the language table
for row in language.itertuples():
    # Check if the language_Id already exists in the table
    cursor.execute('SELECT COUNT(*) FROM Language WHERE language_Id = ?', (row[1],))  # row[1] for language_Id
    count = cursor.fetchone()[0]
    
    if count == 0:
        # Insert the record if it doesn't exist
        cursor.execute('INSERT INTO Language (language_Id, language_name) VALUES (?, ?)', (row[1], row[2]))  # row[1] for language_Id, row[2] for language_name

connection.commit()
```

```python
# Insert data into the Owner table
for row in owner.itertuples():
    # Check if the owner_Id already exists in the table
    cursor.execute('SELECT COUNT(*) FROM Owner WHERE owner_Id = ?', (row.owner_Id,))
    count = cursor.fetchone()[0]
    
    if count == 0:
        # Insert the record if it doesn't exist
        cursor.execute('INSERT INTO Owner (owner_Id, owner_name) VALUES (?, ?)', (row.owner_Id, row.owner))

connection.commit()
```

```python
# Convert 'created_at' column to datetime format
repository['created_at'] = pd.to_datetime(repository['created_at'])

# Insert data into the Repository table
for row in repository.itertuples():
    # Check if the repository_Id already exists in the table
    cursor.execute('SELECT COUNT(*) FROM Repository WHERE repository_Id = ?', (row.repository_Id,))
    count = cursor.fetchone()[0]
    
    if count == 0:
        # Insert the record if it doesn't exist
        cursor.execute('INSERT INTO Repository (repository_Id, owner_Id, repository_name, repository_url, created_at, description, topics, language_Id, stars, forks, open_issues) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)',
                       (row.repository_Id, row.owner_Id, row.repository, row.repository_url, row.created_at, row.description, row.topics, row.language_Id, row.stars, row.forks, row.open_issues))

connection.commit()
```

```python
# Insert data into the Repository_Contributor table
for row in repository_Contributor.itertuples():
    # Check if the combination of 'repository_Id' and 'contributor_Id' already exists in the table
    cursor.execute('SELECT COUNT(*) FROM Repository_Contributor WHERE repository_Id = ? AND contributor_Id = ?', (row.repository_Id, row.contributor_Id))
    count = cursor.fetchone()[0]
    
    if count == 0:
        # Insert the record if it doesn't exist
        cursor.execute('INSERT INTO Repository_Contributor (repository_Id, contributor_Id, contributions_count) VALUES (?, ?, ?)', (row.repository_Id, row.contributor_Id, row.contributions_count))

connection.commit()
```

```python
import time
```

```python
# Exemple de requte  tester avant indexation
query_avant_index = 'SELECT * FROM Repository WHERE repository_name = ?'
valeur_recherche = 'CVE-2023-2002'

# Mesurer le temps d'excution avant indexation
debut_avant_index = time.time()
cursor.execute(query_avant_index, (valeur_recherche,))
resultat_avant_index = cursor.fetchall()
fin_avant_index = time.time()
temps_execution_avant_index = fin_avant_index - debut_avant_index

# Afficher les rsultats
print(f"Temps d'excution avant indexation: {temps_execution_avant_index} secondes")
```

```python
# Crer l'index sur la colonne 'repository_name' de la table 'Repository'
cursor.execute('CREATE INDEX idx_repository_name ON Repository (repository_name);')
connection.commit()
```

```python
# Exemple de requte  tester aprs indexation
query_apres_index = 'SELECT * FROM Repository WHERE repository_name = ?'
valeur_recherche = 'practical-llms'

# Mesurer le temps d'excution aprs indexation
debut_apres_index = time.time()
cursor.execute(query_apres_index, (valeur_recherche,))
resultat_apres_index = cursor.fetchall()
fin_apres_index = time.time()
temps_execution_apres_index = fin_apres_index - debut_apres_index

# Afficher les rsultats
print(f"Temps d'excution aprs indexation: {temps_execution_apres_index} secondes")
```
